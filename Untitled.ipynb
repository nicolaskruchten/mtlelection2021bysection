```python
import pandas as pd
import plotly.express as px
```

```python
px.set_mapbox_access_token(open(".mapbox_token").read())
```

```python
df = pd.read_csv("data.csv")
```

```python
df["NOM_SECTION"] = df.apply(lambda row: str(row["DistrictID"]).zfill(3) +"-"+ str(row["Bureau"]).replace("a","").replace("b","").zfill(3), axis=1)
```

```python
c = ["CODERRE Denis", "PLANTE Valérie"]
df2 = (df.query("Candidat in @c")[["NOM_SECTION", "Candidat", "Votes"]]
       .pivot_table(index="NOM_SECTION", columns="Candidat", values="Votes", aggfunc=sum)
      .reset_index()
      .rename({c[0]: "Denis Coderre", c[1]: "Valérie Plante"}, axis=1))

df2["Winner"] = df2.apply(lambda row: "Denis Coderre" if row["Denis Coderre"]>row["Valérie Plante"] else "Valérie Plante", axis=1)
df2
```
